Valeurs venant du google sheets, retranscrites en csv pour les transformer en dictionnaire de listes ici. J'ai enlevé toutes les valeurs 'X' pour ne pas que le code perde du temps avec ça. (Les valeurs sont en minutes)

```python
import csv

def affiche(montableau): #fonction pour afficher le csv
    with open(montableau, newline='',encoding='utf-8') as f:#"open" pour ouvrir le fichier
        reader = csv.reader(f)
        for ligne in reader:#boucle pour afficher chaque ligne
            print(ligne)

affiche("Trains.csv")
```

```python
temps_train = {}  # Dictionnaire qui contiendra les données
with open("Trains.csv", "r", encoding="utf-8-sig") as fichier:  #"open" pour ouvrir le fichier
    entetes = fichier.readline().strip() #on recupère les valeurs de la première ligne
    cles = entetes.split(",")  # Les clés des dictionnaires sont les en-têtes des colonnes
    for ligne in fichier: # pour chaque ligne du fichier
        valeurs = ligne.split(",")  # Les valeurs des dictionnaires sont les champs séparés par des virgules
        entree = []  # Une entrée représente une ligne dans la table sous forme de liste de tuples
        for i in range(len(cles)): # pour i allant de 0 à la longueur de la ligne
            cle = cles[i] # on prend la valeur en haut de la colonne
            valeur = valeurs[i] # et la valeur correspondant à la distance entre cle et i
            distance_ville = (cle, valeur) # on met ces deux valeurs dans un tuple
            entree.append(distance_ville)  # Ajoutez le tuple (clé, valeur) à la liste
        temps_train[valeurs[0]] = entree  # Ajoutez la liste de tuples à "temps_train" avec la première valeur comme identifiant

temps_train
```

```python
#code nettoyage du dico
for cle in temps_train: #pour chaque cle du dictionnaire temps_train
    i = 0 #on initialise i a 0
    while i < len(temps_train[cle]): #tant que i n'a pas fini de parcourir la longueur de la liste
        if temps_train[cle][i][1] == 'X' or temps_train[cle][i][1] == 'X\n' or temps_train[cle][i][0] == 'VILLES': #si il y a ces valeurs dans les tuples
            del temps_train[cle][i] #on enlève le tuple
        else :
            temps_train[cle][i] = (temps_train[cle][i][0], int(temps_train[cle][i][1])) #je met la distance en integer
            i += 1 #on ajoute 1 à i

print(temps_train)
```

```python
def temps_minute_to_hm(temps):
  """Convertir minutes en heure/minutes"""
  a = temps//60
  b = temps%60
  return (f"{a} heures et {b} minutes")
```

```python
import math
# la valeur infinie s'utilise comme ça : math.inf

# ------------------------------Méthode :---------------------------------------

class Graph:

    #pour l'instant j'ai fait un init avec le graph que j'ai déja écrit au dessus
    def __init__(self,graphique):
        self.graphique=graphique
        self.indice_clé = {}  # Dictionnaire pour lier les clés aux indices

        # Créer un dictionnaire qui lie les clés aux indices
        for i, cle in enumerate(self.graphique.keys()):
            self.indice_clé[cle] = i
            #{ville1:0,ville2:1...}

    #juste pour être sûr que ça a fonctionné
    def __repr__(self):
        return f'la liaison des indices et des clées est : {self.indice_clé}, les données de ce graphes sont : {self.graphique}'

    def dijkstra(self, départ:str, arrivé:str):
        """
        Cette fonction crée un tableau, récupère les informations du graphique
        et applique Dijkstra en renvoyant une liste
        exemple:
        [('ville_dep', 0, None), ('ville2', 2, 'ville_dep'), ('ville3', 3, 'ville2'), ('ville4', 7, 'ville3')]
        """
        #INITIALISATIONS
        global tableau
        ville_pré=None
        nbr_clé = len(self.graphique)
        tableau = creer_liste_vide(nbr_clé)
        liste=[]
        tableau[0][self.indice_clé[départ]] = (départ, 0, None, False)

        for i in range(nbr_clé):

           #CALCULS

           # Stocker le résultat de comparer dans une variable
           comparaison = comparer(tableau)
           # Utiliser la variable pour éviter d'appeler comparer plusieurs fois
           ville_courante = comparaison[0]
           indice_courant = comparaison[3]#colomne
           ville_pré = comparaison[1]

           #on cherche la distance de la ville où nous sommes
           #il y a un try car le premier tour est un None ce qui bloquerait le programme
           try:
               #on prend la distance de la ville précédente
               dist=trouve_dist(ville_pré)
               #on regarde dans quelle ville on est par rapport à la ville précédente
               for voisin,distance in self.graphique[ville_pré]:
                if voisin==ville_courante:
                  #puis on rajoute cette distance à la distance de la ville précédente pour avoir la nouvelle
                  dist=dist+distance
           except:
              #on initialise dist (l'except ne sera fait que au premier tour)
              dist=0

           #ACTUALISATION
           #ne prendra plus en compte cette ville dans la fonction comparer
           ville_visit(indice_courant)
           #on met les infos de la ville_courante sur la nouvelle ligne
           tableau[i][indice_courant] = (ville_courante, dist, ville_pré, True)

           #on complète la liste finale
           liste.append((ville_courante,round(dist,2),ville_pré))

           #pas besoin de continuer le programme si on est déjà arrivé à la ville cible
           if ville_courante==arrivé:
              break

           #actualisation des villes accessibles
           for voisin, distance in self.graphique[ville_courante]:
                indice_voisin = self.indice_clé[voisin]
                if tableau[indice_courant][indice_voisin][3] is False:
                    #on trouve la distance totale pour ce voisin
                    distance_totale = tableau[i][indice_courant][1] + distance
                    tableau[indice_courant][indice_voisin]=(trouver_cle_par_valeur(self.indice_clé,indice_voisin),distance_totale,ville_courante,False)


        return liste

    def trouve_le_chemin_le_plus_court (self,départ:str,arrivé:str):
        """
        cette fonction est juste le résultat final (en utilisant info_finales) de la liste sortie grâce à Dijkstra.
        """
        return info_finales(self.dijkstra(départ,arrivé),arrivé)



# -----------------------------Fonctions :--------------------------------------

def comparer(liste_de_listes):
    """
    compare toutes les valeurs du tableau vérifie qu'elle n'est pas déja confirmée et trouve la ville la plus proche sa ville précédente, et son indice colonne/ligne
    """
    valeur_minimale = math.inf
    ville_min = None
    ville_pré = None

    ligne=0
    colomne=0

    lig=0
    col=0

    for sous_liste in liste_de_listes:

        for tupple in sous_liste:

            # Vérifier si le nœud a été visité et si la distance est plus petite
            if tupple[3] is False and tupple[1] < valeur_minimale:
                valeur_minimale = tupple[1]
                ville_min = tupple[0]
                ville_pré = tupple[2]
                lig=ligne
                col=colomne
            colomne+=1
        ligne+=1
        colomne=0

    return [ville_min,ville_pré,lig,col]

def trouver_cle_par_valeur(dictionnaire, valeur_recherchee):
    for cle, valeur in dictionnaire.items():
        if valeur == valeur_recherchee:
            return cle
    return None

def creer_liste_vide(n):
    """
    initialise un tableau de longueur/largeur = n
    """
    liste_resultat = []

    for _ in range(n):
        sous_liste = [(None, math.inf, None, False)] * n
        liste_resultat.append(sous_liste)

    return liste_resultat

def ville_visit(indice):
    global tableau
    """Met un True sur toutes les lignes d'une ville visitée"""
    for i in range(len(tableau)):
        tableau[i][indice]=(tableau[i][indice][0],tableau[i][indice][1],tableau[i][indice][2],"autre")
    #pourquoi c'est si compliqué ?
    #les tupples sont immuables/non modifiables, si je veux le modifier je dois en recréer un autre en récupérant les valeurs a conserver + ajouter les nouvelles


def trouve_dist (ville):
    """
    On donne le nom de la ville et elle sort sa distance depuis la ville de départ
    """
    global tableau
    for i in tableau:
      for j in i:
        if j[3]==True and j[0]==ville:#si la ville est vérifiée et que c'est la bonne on donne sa distance
          return j[1]


#je vais essayer de faire la dernière fonction celle qui sort les valeurs finales à partir de la liste
def info_finales (liste,arrivé):
    """
    Cette fonction permet de sortir la durée et le chemin empreinté de liste fournie,
    Elle fonctionne en partant de la ville d'arrivée et remonte jusqu'à la ville de départ.
    """
    #liste=[(ville_dep,0,None),(ville2,2,ville_dep),(ville3,3,ville2),(ville4,7,ville3)] arrivé = ville4
    chemin=[arrivé]
    a=1
    while a==1: #boucle qu'on arrète quand a=0
        for i in range(len(liste)):
            #durée
            if liste[i][0]==arrivé:
                durée=liste[i][1]
            #chemin
            if liste[i][0] == chemin[0]:
                if liste[i][2]!= None:
                    chemin.insert(0,liste[i][2])#on insert la valeur liste[i][2] (ville précédente) à l'emplacement 0 de la liste
                else :
                    a=0 #on casse la boucle quand la ville précédente est none
    return f'Les villes empruntées pour passer via le chemin le plus court sont {chemin} pour durée totale de {temps_minute_to_hm(durée)} .'

```
